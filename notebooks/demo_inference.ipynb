```python
import torch
from deepfake_detection.deepfake_detection import DeepfakeDetectionSystem

# Load the working artefact
model = DeepfakeDetectionSystem()
model.load_state_dict(torch.load("../models/checkpoints/final_model.pth"))
model.eval()

def run_inference(video_tensor, audio_tensor):
    with torch.no_grad():
        final_score, v_score, a_score = model(video_tensor, audio_tensor)
        label = "FAKE" if final_score > 0.5 else "REAL"
        print(f"System Output: {label} (Confidence: {final_score.item():.4f})")
        print(f"Video Branch Score: {v_score.item():.4f}")
        print(f"Audio Branch Score: {a_score.item():.4f}")

# Placeholder for actual data loading
print("Ready for Demonstration Video.")
```

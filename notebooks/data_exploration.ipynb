```python
import cv2
import librosa
import numpy as np
import matplotlib.pyplot as plt

# The sys.path modification is removed, as the package should be discoverable via pip install -e .
# import sys
# import os
# repo_root = os.getcwd()
# if repo_root not in sys.path:
#     sys.path.insert(0, repo_root)

from deepfack_project.deepfake_detection.deepfake_detection import DeepfakeVideoDataset

# Load a sample pair
video_sample = "../data/raw/sample_video.mp4"
audio_sample = "../data/raw/sample_audio.wav"

# 1. Video Preprocessing: Frame Extraction
cap = cv2.VideoCapture(video_sample)
ret, frame = cap.read()
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title("Extracted Frame (Preprocessing Step)")
plt.show()

# 2. Audio Preprocessing: 128-bin Mel Spectrogram (Matches Part 1)
y, sr = librosa.load(audio_sample, sr=16000)
mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(mel, ref=np.max), y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('128-bin Mel Spectrogram')
plt.show()
```
